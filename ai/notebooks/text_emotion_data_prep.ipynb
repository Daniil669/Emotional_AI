```python
from datasets import load_dataset
import re, random

def normalize_text(text):
  text = text.lower()
  text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
  return text.strip()

dataset = load_dataset("go_emotions", "simplified")
dataset = dataset.map(lambda d: {'text': normalize_text(d['text'])})

print("\nRandom Samples from Train Set:")
for _ in range(5):
    sample = dataset['train'][random.randint(0, len(dataset['train']) - 1)]
    print(f"Text: {sample['text']}")
    print(f"Labels: {sample['labels']}\n")

dataset = dataset['train'].train_test_split(test_size=0.1)
val_test = dataset['test'].train_test_split(test_size=0.5)
dataset['validation'] = val_test['train']
dataset['test'] = val_test['test']

dataset.save_to_disk('/content/drive/MyDrive/emotional_ai/data/text/goemotions')
print("Dataset saved to Google Drive")
```

```python
from datasets import load_dataset
import re, random

def normalize_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text.strip()

ds = load_dataset("go_emotions", "simplified")

ds = ds.map(lambda d: {'text': normalize_text(d['text'])})

label_names = ds['train'].features['labels'].feature.names

print("\nRandom Samples from TRAIN:")
for _ in range(5):
    sample = ds['train'][random.randint(0, len(ds['train']) - 1)]
    readable = [label_names[i] for i in sample['labels']]
    print(f"Text: {sample['text']}")
    print(f"Labels: {readable}\n")

ds.save_to_disk('/content/drive/MyDrive/emotional_ai/data/text/goemotions')
print("Dataset (train/validation/test) saved to Drive")
```
